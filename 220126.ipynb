```python
df.drop('Unnamed: 0',axis=1,inplace=True) # to remove column named 'Unnamed: 0'
```

```python
df.head()
```

```python
x=df.drop('Sales',axis=1)
```

```python
y=df.Sales
```

```python
df.head()
```

```python
# simple linear regression for Tv
```

```python
# step1 independent and dependent variable creation
x=df[['TV']]  #x variable contains one future
y=df.Sales   #B IS TARGERT
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
print(len(x_test))
print(len(y_train))
```

```python
# Step 3 Creating model
from sklearn.linear_model import LinearRegression
LR=LinearRegression() # object creation
LR.fit(x_train,y_train) # training of linear regression
y_predict=LR.predict(x_test)
```

```python
from sklearn.metrics import (r2_score, mean_squared_error, mean_absolute_error,
root_mean_squared_error)
r2score=r2_score(y_test,y_predict)
r2score
```

```python
print("MSE:",mean_squared_error(y_test,y_predict))
print("MAE:",mean_absolute_error(y_test,y_predict))
print("RMSE:",root_mean_squared_error(y_test,y_predict))
```
